```python
import pandas as pd
```

```python
amazon_data = pd.read_csv("amazon_data.csv")
amazon_data
```

```python
un_needed_column = ['price (including used books)', 'dimensions', 'weight', 'language', 'publisher', 'ISBN_13', 'link', 'complete_link']
amazon_data.drop(un_needed_column, axis = 1, inplace = True)
```

```python
# number of data that are null
amazon_data.isnull().sum()
```

```python
# number of data that are not null
amazon_data.notnull().sum()
```

```python
amazon_data['n_reviews'].fillna(0, inplace = True)
amazon_data
```

```python
amazon_data['avg_reviews'].fillna(0, inplace = True)
amazon_data
```

```python
amazon_data['star5'].fillna(0, inplace = True)
amazon_data['star4'].fillna(0, inplace = True)
amazon_data['star3'].fillna(0, inplace = True)
amazon_data['star2'].fillna(0, inplace = True)
amazon_data['star1'].fillna(0, inplace = True)
amazon_data
```

```python
amazon_data.dropna(0, inplace = True)
```

```python
# reset index to 0
amazon_data.reset_index(drop = True, inplace = True)
```

## converting of objects data to float

```python
amazon_data["star1"] = amazon_data ["star1"].str.rstrip("%").astype(float)/100
```

```python
amazon_data["star2"] = amazon_data ["star2"].str.rstrip("%").astype(float)/100
```

```python
amazon_data["star3"] = amazon_data ["star3"].str.rstrip("%").astype(float)/100
```

```python
amazon_data["star4"] = amazon_data ["star4"].str.rstrip("%").astype(float)/100
```

```python
amazon_data["star5"] = amazon_data ["star5"].str.rstrip("%").astype(float)/100
```

## convert pages from strings to integer

```python
amazon_data["pages"] = pd.to_numeric(amazon_data.pages, errors = 'coerce').fillna(0).astype(int)
amazon_data
```
